### Plotting the 'Height' and 'Weight' distributions

```python
plt.figure(figsize=(20,9))
plt.subplot(1, 2, 1)  
ax = sns.histplot(data = fifa, x='Height (in cm)', kde= True, linewidth=0.6)
ax.set_title('Height (in cm) Distribution in FIFA 21', fontsize=26, loc='left')
ax.set_ylabel('Count', fontsize=20)
ax.set_xlabel('Height (in cm)', fontsize=20)
ax.tick_params(axis='x', labelsize=20)
ax.tick_params(axis='y', labelsize=20)
ax=ax

plt.subplot(1, 2, 2)
ax2 = sns.histplot(data = fifa, x='Weight (in kg)', kde = True,linewidth=0.6)
ax2.set_title('Weight (in kg) Distribution in FIFA 21', fontsize=26, loc='left')
ax2.set_ylabel('Count', fontsize=20)
ax2.set_xlabel('Weight (in Kg)', fontsize=20)
ax2.tick_params(axis='x', labelsize=20)
ax2.tick_params(axis='y', labelsize=20)
ax2=ax2

plt.tight_layout()
```

As we can see, in both variables, they almost follow a normal distribution. This is to be expected, as most football players must have a strict diet and tend to do a lot of physical exercise (that said in relation to the weight of the players). Now regarding their height, the vast majority of them are between 175 and 190 cm.

### Plotting the 'Weak Foot' and 'Skill Move' distributions

```python
plt.figure(figsize=(20,9))
plt.subplot(1, 2, 1)  
ax = sns.histplot(data = fifa, x='W/F', binwidth=0.4)
ax.set_title('Weak Foot stars Distribution in FIFA 21', fontsize=26, loc='left')
ax.set_ylabel('Number of players', fontsize=20)
ax.set_xlabel('Weak Foot stars', fontsize=20)
ax.set_xticks([1, 2, 3, 4, 5])
ax.tick_params(axis='x', labelsize=20)
ax.tick_params(axis='y', labelsize=20)
ax=ax

plt.subplot(1, 2, 2)
ax2 = sns.histplot(data = fifa, x='SM',binwidth=0.4)
ax2.set_title('Skill Moves stars Distribution in FIFA 21', fontsize=26, loc='left')
ax2.set_ylabel('Number of players', fontsize=20)
ax2.set_xlabel('Skill Moves stars', fontsize=20)
ax2.set_xticks([1, 2, 3, 4, 5])
ax2.tick_params(axis='x', labelsize=20)
ax2.tick_params(axis='y', labelsize=20)
ax2=ax2

plt.tight_layout()
```

```python
five_starswf = len(fifa[fifa['W/F'] == 5])
five_starssm = len(fifa[fifa['SM'] == 5])
#Decided to print it because we are not able to see it clearly in the graphs
print(f"The number of players with 5 stars of Weak Foot is {five_starswf}, thats equivalent to {round((five_starswf / len(fifa['W/F'])) * 100, 2)}% of the dataset")
print(f"The number of players with 5 stars of Skill Move is {five_starssm}', thats equivalent to {round((five_starssm / len(fifa['SM'])) * 100, 2)}% of the dataset")
```
